```python
data = randn(100)
```

```python
sns.distplot(data,bins = 25)
```

```python
sns.distplot(data,bins = 10,rug = False,hist = False)
```

```python
sns.distplot(data,bins = 25,
            kde_kws  = {'color': 'red', 'label': 'KDE'},
            hist_kws = {'color': 'blue', 'label': ' Hist'})
```
