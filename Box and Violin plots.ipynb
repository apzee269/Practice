```python
import numpy as np
import pandas as pd
from numpy.random import randn

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data1 = randn(100)
data2 = randn(100)
```

```python
plt.boxplot([data1,data2], patch_artist = True, labels = ['Data 1', 'Data 2'])
plt.show()
```

```python
plt.boxplot([data1,data2],whis = np.inf, patch_artist = True)  #to take outliers into consideration
plt.show()
```

```python
plt.boxplot(data1, vert = False,patch_artist = True, whis = np.inf)
plt.show()
```

```python
#Normal distribution

data1 = stats.norm(0,5).rvs(100)

#Gamma Distribution

data2 = np.concatenate([stats.gamma(5).rvs(50)-1,-1*stats.gamma(5).rvs(50)])

plt.boxplot([data1,data2],patch_artist = True)
plt.show()
```

```python
plt.violinplot([data1,data2])
plt.show()
```

```python
sns.violinplot(data1,bw = 0.01)
```

```python
sns.violinplot(data2,inner = 'stick')
```
