```python
import numpy as np
import pandas as pd

from pandas import DataFrame,Series
```

```python
dframe1 = DataFrame({'key': ['x','z','y','z','x','x'],'Data' : np.arange(6)})
```

```python
dframe1
```

```python
dframe2 = DataFrame({'key': ['q','y','z'], 'Data2': [1,2,3]})
```

```python
dframe2
```

```python
pd.merge(dframe1,dframe2)
```

```python
pd.merge(dframe1,dframe2 , how = 'left')
```

```python
pd.merge(dframe1,dframe2 , how = 'outer') #to pass union of both data frames
```

```python
df = DataFrame({'Key1': ['MH','MP','UP'],
               'Key2': ['one','two','one'],
               'dataf':[10,20,30]})
```

```python
df
```

```python
df2= DataFrame({'Key1': ['MH','MH','MP','MP'],
               'Key2': ['one','one','one','two'],
               'dataf2': [40,50,60,70]})
```

```python
df2
```

```python
pd.merge(df,df2,on =['Key1'],how = 'outer')
```
